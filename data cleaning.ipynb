## *Handling missing values*

```python
# check missin value

df.isnull().sum()
```

```python
# Remove missng values from customer ID column , coz customer Id must be unique

df.dropna(subset = ['Customer_ID'],inplace = True)

# check values remove or not of Customer_ID
df.isnull().sum()
```

```python
# now rows are decrease from 10k to 9k

df
```

```python
# ab ham imputation perform krege on numeric column , remove krege 25 yrs, yrs26 ,etc entries.. ham inmese only num ie, 25,26 ye extract krege
# 34.0 years aise values remove krna h or only unmese numnber extract krna h ie, 34
# df['Age']

df['Age'].unique()
```

```python
# cleans the Age column by extracting only numeric values and removing text like "years" or "yrs" using regular expressions.
import re
def extract_age(age):
    age_num = re.findall('[0-9]+', str(age))
    if len(age_num) > 0:
        return age_num[0]
    else :
        return age
df['Age'] = df['Age'].apply(lambda x : extract_age(x))

df['Age']
```

```python
# check age after extractig numbers(age) from text then Age column looks like
df['Age'].unique()
```

```python
# now handle 'nan years' this value inside Age column
# it only returns only valid age with no 'nan years' value in Age column
df_age = df[df['Age'] != 'nan years']['Age'] 
df_age
```

```python
# Now calculate median age 
# it remove nan value from df_age var temprorily and covert its type into int for calculating median.
# we convert its type coz we are calculating mdian so we need int values and its orginal type is object
    
age_median = int(df_age.dropna().astype('int64').median())
age_median
```

```python
# now fill 'nan years' with median 43 permenetly

df.replace('nan years',age_median,inplace = True)
```

```python
# and also fill value nan with median 43 permenetly

df['Age'].fillna(age_median)
```

```python
# check Age column after removeing nan and 'nan year'
# here 250 is very big age so its a outlier

df['Age'].unique()
```

```python
df.info()
```

```python
# now ham Purchase_Amount col me k missing values ko handle krte h
# isko direct median of Purchase_Amount se fill krdege missing value ko coz isme Purchase_Amount col ka DT yeh alredy numeric h not an object so

df['Purchase_Amount'].fillna(df['Purchase_Amount'].median())
```

```python
# check Purchase_Amount column after removeing missing values

df.isnull().sum()
```

```python
# now ham Feedback_Score col me k missing values ko handle krte h
# isko direct mode of  Feedback_Score se fill krdege missing value ko coz isme  Feedback_Score col ka DT yeh alredy numeric as well as yeh decrete value h

df['Feedback_Score'].fillna(df['Feedback_Score'].mode()[0])
```

```python
# check Feedback_Score after removeing missing values

df.isnull().sum()
```

```python
# now handle missing values of cols Gender,City,Country in one go coz yeh tino columns ka DT object h means string h , so ham isme mode apply krege

for col in ['Gender','City','Country']:
    df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
# check 'Gender','City','Country' after removeing missing values

df.isnull().sum()
```

```python
# now ab ham Last_Purchase_Date col ke missing values ko handle krege with forward fill

df['Last_Purchase_Date'].ffill()
df.isnull().sum()
```

## *Fixing Inconsistent Formatting*

female,
F,
male,
M,
FEMALe,etc

fix this formate into onle male and female

```python
# check incosistence formating in Gender col

df['Gender'].unique()
```

```python
# now convert them into to lower case - temporiry

df['Gender'].str.lower()
```

```python
# now check unique values after converting lower - temprorily

df['Gender'].str.lower().unique()
```

```python
# now remove white spaces from Gender col

df['Gender'].str.lower().str.strip()
```

```python
# now check unique values after removeing white space and converting it into lower case - temprory

df['Gender'] = df['Gender'].str.lower().str.strip()
df['Gender']
```

```python
# now convert m and f into male and female - permently

df['Gender'].replace({
    'm':'male',
    'f':'female'
}, inplace = True)
```

```python
# now check unique values after applying all filters
df['Gender'].unique()
```

```python
# now formate the City col

df['City'].unique()
```

```python
# now remove white spaces
df['City'].str.strip()
```

```python
# now capatalize start word of letter and remove whitespae

df['City'] = df['City'].str.strip().str.capitalize()
```

```python
# now check unique values after applying all filters

df['City'].unique()
```

```python
# Now apply formating in Country col
# ccheck unique values in Country col

df['Country'].unique()
```

```python
# now replace all wrong formate text into write formate ie, India

df['Country'] = df['Country'].replace({
    'india' : 'India',
    'InDia' : 'India',
    'IND'   : 'India'
})
```
